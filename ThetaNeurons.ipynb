```python
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib osx
plt.style.use('seaborn')
```

```python
def f(thetas, t, eta, kappa, I):

    return 1-np.cos(thetas) + (1+np.cos(thetas))*(eta+kappa*I)

def I_fun(an, thetas, n):
    S = 0
    #print("Thetas = ", thetas)
    for i in range(len(thetas)):
        s = (1-np.cos(thetas[i]))**n
        #print("s = ", s)
        S += s
        
    I = an/len(thetas) * S
    return I

def an_fun(n):
    return 2**n*(factorial(n))**2/factorial(2*n)

def factorial(n):
    fact = 1
    for i in range(1, n+1):
        fact = fact*i
    return(fact)

def euler(f, eta, kappa, y0=1, h=1e-3, N=100):
    y = np.ones((len(y0), N))
    y[:, 0] = y0
    #print("y first =", y)
    t = 0
    n = 2
    for i in range(N-1):
        an = an_fun(n=n)
        #print("an = ", an)
        I = I_fun(an, y[:, i], n=n)
        #print("I = ", I)
        y[:, i+1] = y[:, i] + h*f(thetas=y[:, i], t=t, eta=eta, kappa=kappa, I=I)
        #print("y = ", y)
        t += h
    return(y)

        
        
    
```

```python
M = 30000
he = 1e-2

time = np.linspace(0, M, M)/100


thetas1 = np.array([0, 1, 6, 9, 2])
thetas2 = np.array([1e-4, 1, 6, 9, 2])

A = np.sin(euler(f,  eta=0.2, kappa=-0.75, h=he, y0=thetas1, N=M))[0]
B = np.sin(euler(f, eta=0.2, kappa=-0.75, h=he, y0=thetas2, N=M))[0]
print(thetas1)
print(thetas2)
plt.plot(time, A, linewidth=1)
plt.plot(time, B, linewidth=1)
plt.show()
```
